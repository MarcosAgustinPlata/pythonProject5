```python
import random

class Point:
    def __init__(self, x, y):
        """
        This will be called when instantiating an object
        :param x: the value of x
        :param y: the value of y
        """
        self.x = x
        self.y = y

    def __str__(self):
        """
        This will return the string value used in printing the point
        :return:
        """
        return f"Point({self.x}, {self.y})"

    def __repr__(self):
        """
        This is when the point is in a list or other container
        :return:
        """
        return self.__str__()

    def distance_origin(self):
        """
        Calculate the distance to origin of this point
        :return:
        """
        return (self.x**2 + self.y**2)**0.5

    def __gt__(self, other):
        """
        Define how a point is greater than another
        :param other: the other point to compare against
        :return:
        """
        return self.distance_origin() > other.distance_origin()

    def __eq__(self, other):
        """
        Define when 2 points are equal
        :param other:
        :return:
        """
        return self.distance_origin() == other.distance_origin()


p1 = Point(5, 12)
p2 = Point(9, 6)
print(f"p1=({p1.x}, {p1.y})")
print(f"p2=({p2.x}, {p2.y})")

# Creating a list of random 5 points
points = []  # Initialize an empty list
for _ in range(5):
    points.append(Point(random.randint(-50, 50), random.randint(-50, 50)))

for point in points:
    print(f"Point({point.x}, {point.y})")

# Try to print the first point
print("Printing the first point value:", points[0])
print("All points:", points)

p1 = Point(8, 15)
print(f"Distance to origin of p1={p1.distance_origin()}")
p2 = Point(7, 24)
print(f"Distance to origin of p2={p2.distance_origin()}")
print(f"p1 > p2 = {p1 > p2}, p1 < p2 = {p1 < p2}")

points.sort()
print(f"Largest point in the list is {points[-1]}")  # The largest is the last in ordered list

```

```python
class ColoredPoint:
    COLORS = ["red", "green", "blue", "yellow", "black", "white"]

    def __init__(self, x, y, color):
        self.x = x
        self.y = y
        if color in self.COLORS:
            self.color = color
        else:
            raise ValueError(f"That is an invalid color. Accepted colors are {self.COLORS}")

    def __str__(self):
        return f"ColoredPoint(x={self.x}, y={self.y}, color={self.color})"

class AdvancedColorPoint(ColoredPoint):
    def __init__(self, x, y, color):
        """
        Initialize an AdvancedColorPoint with coordinates and a color.
        :param x: x-coordinate
        :param y: y-coordinate
        :param color: color of the point
        """
        self._x = x
        self._y = y
        if color in self.COLORS:
            self._color = color
        else:
            raise ValueError(f"That is an invalid color. Accepted colors are {self.COLORS}")

    @property
    def x(self):
        return self._x

    @x.setter
    def x(self, value):
        raise Exception("Don't try to hack my point!!")

    @property
    def y(self):
        return self._y

    @property
    def color(self):
        return self._color

    @color.setter
    def color(self, value):
        if value in self.COLORS:
            self._color = value
        else:
            raise ValueError(f"That is an invalid color. Accepted colors are {self.COLORS}")

    def __str__(self):
        return f"AdvancedColorPoint(x={self._x}, y={self._y}, color={self._color})"

p = AdvancedColorPoint(10, 10, "blue")
print(p)

# Let's try to hack it
try:
    p.x = 7
except Exception as e:
    print(e)

print(p)

```

```python
import yfinance as yf
import pandas as pd
import numpy as np

# Descargar la lista de las empresas del S&P 500
table = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")[0]
tickers = table["Symbol"].tolist()

# Fechas de inicio y fin para los datos históricos
start_date = "2023-01-01"
end_date = "2023-12-31"

# Descargar los datos históricos de precios ajustados para las empresas del S&P 500 y el índice S&P 500
price_data = yf.download(tickers, start=start_date, end=end_date, group_by="ticker")
sp500 = yf.download("^GSPC", start=start_date, end=end_date)["Adj Close"]

# Crear un DataFrame para los precios ajustados
adj_close_data = pd.DataFrame({ticker: price_data[ticker]['Adj Close'] for ticker in tickers if 'Adj Close' in price_data[ticker].columns})

# Eliminar las empresas con datos faltantes
adj_close_data = adj_close_data.dropna(axis=1)

# Calcular los retornos diarios
returns_data = adj_close_data.pct_change().dropna()
sp500_returns = sp500.pct_change().dropna()

# Calcular los valores beta para cada empresa
beta_values = {}
for ticker in returns_data.columns:
    try:
        stock_returns = returns_data[ticker]
        cov_matrix = np.cov(stock_returns, sp500_returns)
        beta = cov_matrix[0, 1] / cov_matrix[1, 1]
        beta_values[ticker] = beta
    except Exception as e:
        print(f"Error calculating beta for {ticker}: {e}")

# Encontrar las 5 empresas con los valores beta más altos y más bajos
sorted_beta_values = sorted(beta_values.items(), key=lambda x: x[1], reverse=True)

print("Top 5 S&P 500 stocks with the highest beta in 2023:")
for stock, beta in sorted_beta_values[:5]:
    print(f"{stock}: {beta:.2f}")

print("\nTop 5 S&P 500 stocks with the lowest beta in 2023:")
for stock, beta in sorted_beta_values[-5:]:
    print(f"{stock}: {beta:.2f}")

# Graficar los resultados
import matplotlib.pyplot as plt

# Extraer las empresas y los valores beta
high_beta_stocks = [stock for stock, beta in sorted_beta_values[:5]]
high_beta_values = [beta for stock, beta in sorted_beta_values[:5]]
low_beta_stocks = [stock for stock, beta in sorted_beta_values[-5:]]
low_beta_values = [beta for stock, beta in sorted_beta_values[-5:]]

# Graficar las empresas con los valores beta más altos y más bajos
plt.figure(figsize=(10, 5))
plt.bar(high_beta_stocks, high_beta_values, color='red', label='Highest Beta')
plt.bar(low_beta_stocks, low_beta_values, color='green', label='Lowest Beta')
plt.xlabel('Stocks')
plt.ylabel('Beta Values')
plt.title('Top 5 S&P 500 Stocks with Highest and Lowest Beta in 2023')
plt.legend()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Descargar la lista de las empresas del S&P 500
table = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")[0]
tickers = table["Symbol"].tolist()

# Fechas de inicio y fin para los datos históricos
start_date = "2023-01-01"
end_date = "2023-12-31"

# Descargar los datos históricos de precios ajustados para las empresas del S&P 500 y el índice S&P 500
price_data = yf.download(tickers, start=start_date, end=end_date, group_by="ticker")
sp500 = yf.download("^GSPC", start=start_date, end=end_date)["Adj Close"]

# Crear un DataFrame para los precios ajustados
adj_close_data = pd.DataFrame({ticker: price_data[ticker]['Adj Close'] for ticker in tickers if 'Adj Close' in price_data[ticker].columns})

# Eliminar las empresas con datos faltantes
adj_close_data = adj_close_data.dropna(axis=1)

# Calcular los retornos diarios
returns_data = adj_close_data.pct_change().dropna()
sp500_returns = sp500.pct_change().dropna()

# Crear un DataFrame para almacenar los resultados
results = pd.DataFrame(index=returns_data.columns, columns=['Mean Return', 'Std Dev', 'Correlation'])

# Calcular el promedio, la desviación estándar y la correlación para cada acción
for ticker in returns_data.columns:
    mean_return = returns_data[ticker].mean()
    std_dev = returns_data[ticker].std()
    correlation = returns_data[ticker].corr(sp500_returns)
    results.loc[ticker] = [mean_return, std_dev, correlation]

# Mostrar los resultados
print("Statistical Analysis of S&P 500 Stocks in 2023")
print(results)

# Graficar la correlación de las acciones con el S&P 500
plt.figure(figsize=(10, 5))
plt.bar(results.index, results['Correlation'].astype(float), color='blue')
plt.xlabel('Stocks')
plt.ylabel('Correlation with S&P 500')
plt.title('Correlation of S&P 500 Stocks with S&P 500 in 2023')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas_datareader as pdr
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
from statsmodels.tsa.seasonal import STL
import matplotlib.pyplot as plt

# Cargar y preparar datos
start_date = '2000-01-01'
end_date = '2024-05-01'
data = pdr.get_data_fred('APU000072610', start_date, end_date)
data_series = data.iloc[:, 0]

# STL decomposition
stl = STL(data_series, seasonal=13)
result = stl.fit()

# Extraer componentes
trend = result.trend.dropna().values
seasonal = result.seasonal.dropna().values
real = data_series.dropna().values

# Asegurar que todas las series tengan la misma longitud
min_length = min(len(trend), len(seasonal), len(real))
trend = trend[:min_length]
seasonal = seasonal[:min_length]
real = real[:min_length]

# Escalar datos
scaler = MinMaxScaler()
trend_scaled = scaler.fit_transform(trend.reshape(-1, 1))
seasonal_scaled = scaler.fit_transform(seasonal.reshape(-1, 1))
real_scaled = scaler.fit_transform(real.reshape(-1, 1))

# Preparar datos para la regresión lineal
X = np.hstack([trend_scaled, seasonal_scaled])
X = sm.add_constant(X)  # Agregar constante para el término de intercepto
y = real_scaled

# Crear y entrenar el modelo de regresión lineal
model = sm.OLS(y, X).fit()

# Obtener predicciones
y_pred = model.predict(X)

# Invertir la escala de los datos para graficar
y = scaler.inverse_transform(y)
y_pred = scaler.inverse_transform(y_pred.reshape(-1, 1))

# Graficar resultados
plt.figure(figsize=(12, 6))
plt.plot(range(len(y.flatten())), y.flatten(), label='Real', color='blue')
plt.plot(range(len(y_pred.flatten())), y_pred.flatten(), label='Predicho', color='red')
plt.title('Predicciones vs Reales (Regresión Lineal con Trend y Seasonality)')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend()
plt.show()

# Mostrar resumen del modelo
print(model.summary())
```
